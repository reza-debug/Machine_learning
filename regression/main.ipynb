```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4]).reshape(-1,1)
y= np.array([1,3,5,7])
#x ra reshape konim ke az halate (4,) khare shavad
#mikhaym bedonim agar x=5 beshe y chi mishe
```

```python
plt.plot(x,y,"*")
plt.show()
```

```python
#Model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
#agar x ra reshape nakonim be sorate (4,) ast va khata midahad
```

```python
#y = m* x +c
#Moadeleye Khat
#m=shib
#c=arz az mabdA
m = model.coef_[0]
#coef= zaribhaye x
c = model.intercept_
#intercet= +c

print(m,c)

```

```python
plt.plot(x,y,"*")
plt.plot(x,model.predict(x))
plt.show()
```

```python
result = model.predict([[5]])
result
```
